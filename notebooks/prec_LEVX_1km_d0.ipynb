```python
import pandas as pd

station_all = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/LEVX_1km/input_files/LEVX_20192020.csv",parse_dates=["time"]).drop(columns=["Unnamed: 0"]).set_index("time")
print(station_all)
pd.options.display.max_rows = 999
station_all["wxcodes_o"].value_counts(normalize=True).map(lambda n: '{:.1%}'.format(n))
```

```python
#label precipitation
station = station_all[["wxcodes_o","temp_o"]].astype(str)
station["prec_l"] = "No RA/DZ"
mask = station['wxcodes_o'].str.contains("RA")
station.loc[mask,["prec_l"]] = "RA/DZ"
mask = station['wxcodes_o'].str.contains("DZ")
station.loc[mask,["prec_l"]] = "RA/DZ"
pd.options.display.max_rows = 999
station = station.drop(columns =["temp_o"])
station.sample(100)

```

**Forecast D0**

```python
#ria forecast d0
model1k = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/LEVX_1km/input_files/lat42.22lon-8.63p2R1KmD0.csv",parse_dates=["time"]).set_index("time")
model1k
```

```python
#label meteorological model prec points 0 to 9
model1k["prec0_l"] = ["RA/DZ" if c > 0 else "No RA/DZ" for c in model1k.prec0]
model1k["prec1_l"] = ["RA/DZ" if c > 0 else "No RA/DZ" for c in model1k.prec1]

```

```python
#concat station model
df_all1k = pd.concat([station,model1k],axis=1).dropna()
df_all1k[["prec_l","prec0_l","prec1_l"]].sample(20)
```

```python
print("station\n",df_all1k["prec_l"].value_counts(normalize=True))
print("point 0\n",df_all1k["prec0_l"].value_counts(normalize=True))
print("point 1\n",df_all1k["prec1_l"].value_counts(normalize=True))

```

```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 9))
plt.title("Quality report meteorologic model")
cr = pd.DataFrame(classification_report(df_all1k.prec_l, df_all1k.prec0_l,output_dict=True)).T
sns.heatmap(cr[:-3].drop(columns=["support"]),annot=True,cmap="YlGnBu",fmt='.0%');
```

```python
plt.figure(figsize=(8, 6))
plt.title("Global quality report (meteorologic model)")
sns.heatmap(cr[-3:].drop(columns=["support"]),annot=True,cmap="YlGnBu",fmt='.1%');
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 9))
plt.title("Confusion matrix (meteorologic model)")
global_sc = pd.crosstab(df_all1k.prec_l, df_all1k.prec0_l,margins=True,)
sns.heatmap(global_sc,annot=True,cmap="YlGnBu",fmt='.0f');
```

```python
import math
from scipy.stats import entropy


plt.figure(figsize=(12, 9))
plt.title("Precision and entropy meteorologic model")
column_sc = pd.crosstab(df_all1k.prec_l, df_all1k.prec0_l, margins=True,normalize="columns")
column_sc = column_sc.append(pd.DataFrame(entropy(column_sc,base=2)/(math.log2(column_sc.shape[0])),columns=["entropy/entropy.max"],
                    index=column_sc.columns).T).rename(columns={"All":"Climatology"})
sns.heatmap(column_sc,annot=True,cmap="YlGnBu",fmt='.0%');
```

```python
print("Confusion matrix (meteorologic model)\n")
for l in ["prec0_l","prec1_l"]:
  cm = pd.crosstab(df_all1k.prec_l, df_all1k[l],margins=True,)
  print(cm)
  a = cm.values[0,0]
  b = cm.values[1,0]
  c = cm.values[0,1]
  d = cm.values[1,1]
  hss = 2*(a*d-b*c)/((a+c)*(c+d)+(a+b)*(b+d))
  print("Heidke Skill Score:",round(hss,2))
  print("\n")
```

**ML LGBMClassifier**

```python
from lightgbm.sklearn import LGBMClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

#select x variables
x_var = model1k.drop(columns=["Unnamed: 0","prec0_l","prec1_l"]).columns.to_list()  

# X and Y
Y = df_all1k.prec_l
X = df_all1k[x_var]

# loop ranadom
HSS = []
for t in range (0,10):
  #split variables
  X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.15, )

  # Resample
  X_res, y_res = SMOTE().fit_resample(X_train,y_train)

  ml_model=LGBMClassifier(n_estimators=250).fit(X_res,y_res)

  #Confussion matrix

  print("\nMatriz de Confusión  (Machine Learning)\n")

  cm = pd.crosstab(y_test, ml_model.predict(X_test),margins=True,)
  print(cm)
  a = cm.values[0,0]
  b = cm.values[1,0]
  c = cm.values[0,1]
  d = cm.values[1,1]
  HSS.append(2*(a*d-b*c)/((a+c)*(c+d)+(a+b)*(b+d)))
  print("Heidke Skill Score:",round(HSS[t],2))

  #quality report

  print("\nInforme de calidad desagregado (Machine Learning)\n")
  print(pd.DataFrame(classification_report(y_test, ml_model.predict(X_test),output_dict=True)).T)

print("\nEstadiísticas HSS")
pd.DataFrame(HSS,columns=["HSS"]).describe()
```
